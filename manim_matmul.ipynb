# First attempts at visualising matmuls in manim

```python
from manim import *
import numpy as np
```

```python
%%manim -qm -v WARNING LinearTransform3D

class LinearTransform3D(ThreeDScene):
    def construct(self):
        self.set_camera_orientation(phi=65*DEGREES, theta=30*DEGREES)

        M = np.array([[1., 1., 0.],
                      [0., 1., 0.],
                      [0., 0., 2.]])
        # v0 = np.array([1., 1., 1.])
        v0 = np.array([-0.0737666, 1.2749, 0.107909])

        axes = ThreeDAxes(x_range=[-3, 3, 1], y_range=[-3, 3, 1], z_range=[-3, 3, 1])

        def arrow(end, color):
            return Arrow3D(
                start=np.array([0., 0., 0.]),
                end=np.array(end, dtype=float),
                color=color,
                thickness=0.04,
                height=0.25,
                base_radius=0.1,
            )

        def label(text, tip, color):
            tip = np.array(tip, dtype=float)
            offset = 0.35 * tip / (np.linalg.norm(tip) + 1e-9)
            return Text(text, color=color, weight=BOLD).scale(0.6).move_to(tip + offset)

        i_hat = arrow([1, 0, 0], RED)
        j_hat = arrow([0, 1, 0], GREEN)
        k_hat = arrow([0, 0, 1], BLUE)
        v     = arrow(v0, YELLOW)

        x_lbl = label("x", [1, 0, 0], RED)
        y_lbl = label("y", [0, 1, 0], GREEN)
        z_lbl = label("z", [0, 0, 1], BLUE)

        self.add(axes, i_hat, j_hat, k_hat, v)
        self.add_fixed_orientation_mobjects(x_lbl, y_lbl, z_lbl)
        self.wait(2)

        self.begin_ambient_camera_rotation(rate=0.2)
        
        self.wait(3)

        self.play(
            ReplacementTransform(i_hat, arrow(M @ np.array([1., 0., 0.]), RED)),
            ReplacementTransform(j_hat, arrow(M @ np.array([0., 1., 0.]), GREEN)),
            ReplacementTransform(k_hat, arrow(M @ np.array([0., 0., 1.]), BLUE)),
            ReplacementTransform(v,     arrow(M @ v0, YELLOW)),
            Transform(x_lbl, label("x", M @ np.array([1., 0., 0.]), RED)),
            Transform(y_lbl, label("y", M @ np.array([0., 1., 0.]), GREEN)),
            Transform(z_lbl, label("z", M @ np.array([0., 0., 1.]), BLUE)),
            run_time=4,
        )
        self.wait(2)
        self.stop_ambient_camera_rotation()
        
```
